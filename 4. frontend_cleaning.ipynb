#### Categories

```python
cgr = pd.read_csv('historico.csv', encoding='utf-8')
```

```python
#limpieza inicial
cgr = cgr.drop(['Unnamed: 0','codigo_cgr'], axis=1)
cgr.rename(columns={'divipola':'cod_municipio'}, inplace=True)
cgr['cod_municipio'] = cgr['cod_municipio'].astype('int64')
cgr['cod_municipio'] = cgr['cod_municipio'].astype(str)
```

```python
#concatenar categorias por año

years = [2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]

df = cgr[['cod_municipio','departamento', 'MUNICIPIO','cat_2012']]
df['año'] = '2012'
df.rename(columns={'cat_2012':'categoria'},inplace=True)

for year in years:
  aux_df = cgr[['cod_municipio','departamento', 'MUNICIPIO',f'cat_{year}']]
  aux_df['año'] = year
  aux_df.rename(columns={f'cat_{year}':'categoria'}, inplace=True)
  df = pd.concat([df, aux_df])
```

```python
df['cod_municipio'] = df['cod_municipio'].astype(str)
df['año'] = df['año'].astype(str)
df['key'] = df['cod_municipio'] + '_' + df['año']
```

```python
df.to_csv('categories.csv',encoding='utf-8', index=False)
```
